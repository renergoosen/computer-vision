```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import imutils
from imutils.object_detection import non_max_suppression
```

```python
imPath = r'p4.jpg'
im = cv2.imread(imPath)
```

```python
plt.imshow(im[:,:,::-1])
```

```python
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

```python
(r,w) = hog.detectMultiScale(im,winStride=(2,2), # rectangles and waves, detects at different scales. 2x2 window
                             padding=(16,16),scale=1.5)
```

```python
rects = np.array([[x,y,x+w,y+h] for (x,y,w,h) in r]) # draw rectangle on original image
```

```python
rects_NMS = non_max_suppression(rects,probs=None,overlapThresh=0.6) # non maximally supressed rectangles
```

```python
for (x1,y1,x2,y2) in rects_NMS:
    cv2.rectangle(im,(x1,y1),(x2,y2),(0,0,255),2) # drawing rectangles over image, thickness of rectangle is 2
```
