# RGB Colour Images

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
cim = plt.imread(r'tulips.jpg')
cim.shape
R = cim[:,:,0]
G = cim[:,:,1]
B = cim[:,:,2]
plt.figure(1)

plt.subplot(231) # 2 rows, 3 colums
plt.imshow(cim)
plt.xticks([]) # remove x axis labels
plt.yticks([]) # remove y axis labels

plt.subplot(232) 
plt.imshow(cim)
plt.xticks([]) # remove x axis labels
plt.yticks([]) # remove y axis labels

plt.subplot(233)
plt.imshow(cim)
plt.xticks([]) # remove x axis labels
plt.yticks([]) # remove y axis labels

plt.subplot(234)
plt.imshow(R, cmap = 'gray')
plt.xticks([]) # remove x axis labels
plt.yticks([]) # remove y axis labels
plt.title('Red') #add title

plt.subplot(235)
plt.imshow(G, cmap = 'gray')
plt.xticks([]) # remove x axis labels
plt.yticks([]) # remove y axis labels
plt.title('Green')

plt.subplot(236)
plt.imshow(B, cmap = 'gray')
plt.xticks([]) # remove x axis labels
plt.yticks([]) # remove y axis labels
plt.title('Blue')

# same image copied multiple times with the three colour channels, brightness represents the colour's contribution

plt.show()
```

# OpenCV has BGR ordering

```python
# using open CV:

import numpy as np
import matplotlib.pyplot as plt
import cv2
cim = cv2.imread(r'tulips.jpg')
cim.shape

cim = cim[:,:,::-1] # reversing colour channels to be BGR

R = cim[:,:,0] # Blue
G = cim[:,:,1] # Green
B = cim[:,:,2] # Red
plt.figure(1)

plt.subplot(231) # 2 rows, 3 colums
plt.imshow(cim)
plt.xticks([]) # remove x axis labels
plt.yticks([]) # remove y axis labels

plt.subplot(232) 
plt.imshow(cim)
plt.xticks([]) # remove x axis labels
plt.yticks([]) # remove y axis labels

plt.subplot(233)
plt.imshow(cim)
plt.xticks([]) # remove x axis labels
plt.yticks([]) # remove y axis labels

plt.subplot(234)
plt.imshow(R, cmap = 'gray')
plt.xticks([]) # remove x axis labels
plt.yticks([]) # remove y axis labels
plt.title('Red') #add title

plt.subplot(235)
plt.imshow(G, cmap = 'gray')
plt.xticks([]) # remove x axis labels
plt.yticks([]) # remove y axis labels
plt.title('Green')

plt.subplot(236)
plt.imshow(B, cmap = 'gray')
plt.xticks([]) # remove x axis labels
plt.yticks([]) # remove y axis labels
plt.title('Blue')

# OpenCV by default does not have RGB ordering, it has BGR ordering

plt.show()
```

# Challenge: Contrutct an image in python with RGB and BGR separations

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

rgb = np.zeros((100,150,3), dtype = 'uint8') # create an array of zeros, with 100 rows, 150 colums and 3 channels. Set dtype

rgb [:,0:50,0] = 255 # red is first colour channel (0)
rgb[:,50:100,1] = 255 # green is second colour channel (1)
rgb[:,100:150,2] = 255 # blue is third colour channel (3)


plt.imshow(rgb)
```
