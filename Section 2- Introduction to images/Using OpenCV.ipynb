# Greyscale images in OpenCV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread(r'albert-einstein_gray.jpg', cv2.IMREAD_GRAYSCALE) # ensures OpenCV read in greyscale
cv2.imshow('Gray',img)
cv2.waitKey(0)
cv2.destroyAllWindows
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread(r'albert-einstein_gray.jpg', cv2.IMREAD_GRAYSCALE) # ensures OpenCV read in greyscale
img[500:700, 500:600] = 255
cv2.imshow('Gray',img)
cv2.waitKey(0)
cv2.destroyAllWindows
cv2.imwrite(r'albert-einstein_gray_OpenCV.jpg', img)
```
