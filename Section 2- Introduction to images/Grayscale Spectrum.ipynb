```python
import numpy as np
import matplotlib.pyplot as plt
im = np.arange(256) # 4 colour options (0,1 ,2, 3) with 0 being pure black and 3 being pure white
# number in bracket determines the number of grey scales, starting at 0 which is black
im.shape
im = im[np.newaxis,:]
im.shape
im = np.repeat(im, 100, axis=0) # repeat image 10 times vertically, so 10 rows, each with 4 colours
# number determines how many times the image is repeated vertically
im.shape
im  # have a matrix with 10 rows and 4 colums, with the colums being 0, 1, 2 and 3
plt.imshow(im, cmap = "gray" )  # Plot the image in grey scale
# cmap maps the smallest value in array to black and largest to white
```

```python
# Processing grey scale images from a computer
import matplotlib.pyplot as plt
import matplotlib.image as mp
im = plt.imread(r"albert-einstein_gray.jpg")
im.shape  # shape of matrix
im.dtype  # data type: unsigned integer, 8 bits
plt.imshow(im, cmap = "gray")

im[23, 300]  # Accessing colour code in 24th row and 301st collumn, which is 70
im2 = im.copy()  # copy original image so we can do some more with it
im2[23:100, 40:100] = 255  # start from row 24 and go to row 100 (100 not included), colum 41 to colum 100
im2[300:400, 40:100] = 100
plt.imshow(im2, cmap = "gray")
# There is a white patch and a grey patch, we have set one area selected to white (255) and one to grey (100)
plt.imsave(r"albert-einstein_gray_modified.jpg", im2, cmap = "gray")
```
