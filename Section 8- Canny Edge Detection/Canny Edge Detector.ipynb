```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
imPth = r'cat.jpg'
img = plt.imread(imPth)
```

```python
plt.imshow(img,cmap='gray')
```

```python
E = cv2.Canny(img,250,300)
```

```python
plt.subplot(121)
plt.imshow(img,cmap='gray')
plt.axis('off')
plt.subplot(122)
plt.imshow(E,cmap='gray')
plt.axis('off')
plt.show()
```
