```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_gray = plt.imread(r'albert-einstein_gray.jpg')
img_binary = plt.imread(r'j.png')
img_closed = plt.imread(r'j_closed.png')
img_opened = plt.imread(r'j_opened.png')
```

```python
plt.imshow(img_closed,cmap='gray')
```

```python
K_r = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5)) # rectangular kernel with 5 x 5 dimensions
K_e = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5)) # ellipse
K_c = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5)) # cross
```

```python
TH_c = cv2.morphologyEx(img_opened,cv2.MORPH_TOPHAT,K_c)
```

```python
plt.imshow(TH_c,cmap='gray')
```

```python
BH_c = cv2.morphologyEx(img_closed,cv2.MORPH_BLACKHAT,K_c)
plt.imshow(BH_c,cmap='gray')
```
