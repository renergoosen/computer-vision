## Binary images are images consisting of only black (0) and white (1) pixels

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_gray = plt.imread(r'albert-einstein_gray.jpg')
img_binary = plt.imread(r'j.png')
```

```python
plt.imshow(img_binary)
```

```python
np.unique(img_binary)  # only 2 unique points(1 and 2)
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
# converting to binary image
th = 100 # colour threshold
im2 = img_gray>th # any pixel with value greater than threshold will be white and others will be black
plt.imshow(im2, cmap = 'gray')
```

```python
# inverting the colours
plt.imshow(~(im2),cmap='gray')
```
