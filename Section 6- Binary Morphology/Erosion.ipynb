```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_gray = plt.imread(r'albert-einstein_gray.jpg')
img_binary = plt.imread(r'j.png')
```

```python
K_r = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5)) # rectangular kernel with 5 x 5 dimensions
K_e = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5)) # ellipse
K_c = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5)) # cross
```

```python
# corrosion kernels of different shapes
E_r = cv2.erode(img_binary,K_r)
E_e = cv2.erode(img_binary,K_e)
E_c = cv2.erode(img_binary,K_c)
```

```python
plt.subplot(221)
plt.imshow(img_binary)
plt.subplot(222)
plt.imshow(E_r)
plt.subplot(223)
plt.imshow(E_e)
plt.subplot(224)
plt.imshow(E_c)
```

## cross and elipse seem to be giving the same image so lets see if it actually is the same

```python
print(np.sum((E_r-E_e)**2))
```

#### there are 936 pixels that disagree in each of the images so they are not the same

```python
print(np.sum((E_r - E_c)**2))
```

```python
print(np.sum((E_c - E_e)**2))
```
