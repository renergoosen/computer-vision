```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_gray = plt.imread(r'albert-einstein_gray.jpg')
img_binary = plt.imread(r'j.png')
```

```python
K_r = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5)) # rectangular kernel with 5 x 5 dimensions
K_e = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5)) # ellipse
K_c = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5)) # cross
```

```python
D_r = cv2.dilate(img_binary,K_r)
D_e = cv2.dilate(img_binary,K_e)
D_c = cv2.dilate(img_binary,K_c)
```

```python
plt.subplot(221)
plt.imshow(img_binary)
plt.title("Original")
plt.axis('off')
plt.subplot(222)
plt.imshow(D_r)
plt.title("Dilate Rect")
plt.axis('off')
plt.subplot(223)
plt.imshow(D_e)
plt.title("Dilate Ellipse")
plt.axis('off')
plt.subplot(224)
plt.imshow(D_c)
plt.title("Dilate Cross")
plt.axis('off')
plt.show()
```
