```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import cv2
```

```python
im1 = cv2.imread(r'panorama_1.jpg')
im2 = cv2.imread(r'panorama_2.jpg')
im3 = cv2.imread(r'panorama_3.jpg')
```

```python
plt.subplot(131)
plt.imshow(im1[:,:,::-1])
plt.xticks([])
plt.yticks([])
plt.subplot(132)
plt.imshow(im2[:,:,::-1])
plt.xticks([])
plt.yticks([])
plt.subplot(133)
plt.imshow(im3[:,:,::-1])
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# creating image list
imgs = []
imgs.append(im1)
imgs.append(im2)
imgs.append(im3)
```

```python
M = cv2.Stitcher.create(cv2.Stitcher_PANORAMA)
status,pano = M.stitch(imgs)
```

```python
plt.imshow(pano[:,:,::-1])
plt.xticks([])
plt.yticks([])
```
